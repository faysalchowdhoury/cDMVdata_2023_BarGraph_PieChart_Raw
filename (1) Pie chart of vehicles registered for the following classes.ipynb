```python
import matplotlib.pyplot as plt

# Data
labels = ['Diesel-LDV', 'Diesel-HDV', 'Natural Gas-LDV', 'Natural Gas-HDV', 'Ethanol (E-85)-LDV', 'Ethanol (E-85)-HDV', 'Electricity-LDV', 'Electricity-HDV']
sizes = [588808, 750627, 7677, 25965, 1171644, 47265, 1141466, 2457]

# Create a pie chart
plt.figure(figsize=(10, 10))
wedges, _, autotexts = plt.pie(sizes, labels=labels, autopct=lambda p: f'{p:.1f}% ({int(p * sum(sizes) / 100):,})', startangle=140, shadow=True, colors=plt.cm.Paired.colors)

# Add a title
plt.title('Fuel Type Distribution', fontsize=16)

# Add a legend with fuel name, vehicle number, and vehicle percentage
plt.legend(wedges, labels, title='Fuel Types', loc='upper left', bbox_to_anchor=(0.85, 0, 0.5, 1))

# Annotate the vehicle number and percentage inside each slice
for autotext in autotexts:
    autotext.set_fontsize(12)

# Equal aspect ratio ensures that the pie is drawn as a circle.
plt.axis('equal')

# Show the pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Diesel-LDV', 'Diesel-HDV', 'Natural Gas-LDV', 'Natural Gas-HDV', 'Ethanol (E-85)-LDV', 'Ethanol (E-85)-HDV', 'Electricity-LDV', 'Electricity-HDV']
sizes = [588808, 750627, 7677, 25965, 1171644, 47265, 1141466, 2457]

# Create a pie chart
plt.figure(figsize=(10, 10))
wedges, _, autotexts = plt.pie(sizes, labels=labels, autopct=lambda p: f'{p:.1f}%\n({int(p * sum(sizes) / 100):,})', startangle=140, shadow=True, colors=plt.cm.Paired.colors)

# Add a title
plt.title('Fuel Type Distribution', fontsize=16)

# Add a legend with fuel name, vehicle number, and vehicle percentage
plt.legend(wedges, labels, title='Fuel Types', loc='best')

# Annotate the vehicle number and percentage inside each slice
for autotext in autotexts:
    autotext.set_fontsize(12)

# Equal aspect ratio ensures that the pie is drawn as a circle.
plt.axis('equal')

# Show the pie chart
plt.tight_layout()  # Prevents overlapping of text and ensures the entire plot is visible
plt.show()
```
